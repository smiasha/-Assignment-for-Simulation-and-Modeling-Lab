2.Analyze Matrix A

Using NumPy, compute:

● Shape of A

● Determinant

● Rank

● Eigenvalues

● Inverse (only if determinant ≠ 0)

```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

d1 = 1
d2 = 2

A = np.array([[d1+2, d2+1], [2*d1, d2+2]])

#shape of A
print("shape of matrix A is: ", A.shape)

# determinant
det = np.linalg.det(A)
print("determinant of A: ", det)

# rank
rank = np.linalg.matrix_rank(A)
print("rank of matrix A is: ", rank)

# eigenvalues
eigvals = np.linalg.eigvals(A)
print("eigen values: ", eigvals)

# inverse of matrix A (if exists)
if abs(det) > 0.0001:
    inv = np.linalg.inv(A)
    print("inverse of matrix A is:\n", inv)
else:
    print("determinant is zero, inverse cannot be calculated")
```

4. Re-analyze Matrix B

Compute the same properties as in Step 2.

```python
import numpy as np

d1 = 1
d2 = 2

# Step 1: Original matrix A
A = np.array([[d1+2, d2+1], [2*d1, d2+2]])
print("Original matrix A:\n", A)

# Step 2 (Question 3): Create matrix B by changing ONE element
B = A.copy()
B[0, 0] = B[0, 0] + 1  # Changed element: top-left (row 0, column 0) → added +1
print("\n[Q3] Changed element: A[0,0] (top-left) by adding +1")
print("New matrix B:\n", B)

# Step 3 (Question 4): Re-analyze matrix B
print("\n[Q4] Analysis of matrix B:")

print("shape of matrix B is: ", B.shape)
print("determinant of B: ", np.linalg.det(B))
print("rank of matrix B is: ", np.linalg.matrix_rank(B))
print("eigen values: ", np.linalg.eigvals(B))

det_B = np.linalg.det(B)
if abs(det_B) > 1e-10:
    print("inverse of matrix B is:\n", np.linalg.inv(B))
else:
    print("determinant is zero, inverse does not exist")
```

A = [[3, 2], [2, 3]] B = [[4, 2], [2, 3]]

1.How did the determinant change and why?
Answer:

Determinant increased from 5 → 8 (got bigger).

Why?

Determinant of a 2×2 matrix [[a,b],[c,d]] = a×d – b×c. We increased a (top-left) from 3 → 4: Old: 3×3 – 2×2 = 9 – 4 = 5 New: 4×3 – 2×2 = 12 – 4 = 8

Determinant measures "how much space the matrix stretches". By increasing one element, we stretched the space more → determinant increased.

2.Did the rank change? Explain.

Answer:

No change — both A and B have rank = 2. Why? Rank = number of independent rows/columns A 2×2 matrix has full rank (2) if its rows are not multiples of each other In both A and B: Row 1 = [3,2] or [4,2] Row 2 = [2,3] Neither row is a multiple of the other → both are independent

Changing one number slightly didn't make the rows "line up" or become dependent. So rank stayed full (2).

3.How did the eigenvalues respond to the value change?

Answer:

Both eigenvalues increased slightly: λ₁: 5 → ~5.56 (+0.56) λ₂: 1 → ~1.44 (+0.44)

Why?

Eigenvalues show "how much the matrix stretches vectors in special directions". When we increased the top-left element: The matrix became "stronger" at stretching space So both stretching factors (eigenvalues) increased

4.Is B easier or harder to invert than A? Why?

Answer:

B is slightly easier to invert than A.

Why?

A matrix is "easier to invert" when its determinant is larger (farther from zero) A: det = 5 B: det = 8 (bigger!) Also, eigenvalues of B (5.56, 1.44) are farther from zero than A's (5, 1)

Think of determinant as "stability score": det = 0 → impossible to invert (broken) det = small number → unstable, sensitive to errors det = large number → stable, easy to invert Since B's determinant (8) > A's (5), B is more stable → easier to invert.
